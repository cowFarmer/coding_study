# 파이썬 자료구조와 알고리즘 for beginner
### Chapter 10 재귀 호출

```python
import random

def arySum(arr, n):
    if n <= 0:
        return arr[0]
    return arySum(arr, n-1)+arr[n]

ary = [random.randint(0,255) for _ in range(random.randint(5,10))]
print(ary)
print(ary[0])
print(len(ary)-1)
print(arySum(ary, len(ary)-1))
```

피보나치

```python
def fibo(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)

print('0 1', end = ' ')
for i in range(2, 20):
    print(fibo(i), end = ' ')
```
