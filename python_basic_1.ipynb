# 4-6. 퀴즈풀어보기
사이트의 주소를 받으면 비밀번호 만들기   
예시 사이트 https://www.naver.com   
- 규칙1. naver.com만 사용하기
- 규칙2. 처음 만나는 . 이후 부분은 제외하기
- 규칙3. 남은 글자 중 처음 세자리 + 글자 수 + 글자 내 'e' 개수 + "!" 로 구성하기

```python
def make_password(site):
    # 규칙1
    https = site.find("https://")
    https = site.find("//")
    site = site[https+2:] # www.naver.com

    # 규칙2
    find_dot = site.find(".")
    site = site[find_dot+1:] # naver.com
    find_dot = site.find(".")
    site = site[:find_dot] # naver

    # 규칙3
    cut_site = site[:3]
    site_len = len(site)
    e_word_len = site.count("e")
    password = f"{cut_site}{site_len}{e_word_len}!"
    return password

naver = "https://www.naver.com"
make_password(naver)
```

```python
def make_password(url):
    my_str = url.replace("https://www.", "") # 규칙1
    my_str = my_str[:my_str.index(".")] # 규칙2
    password = my_str[:3]+str(len(my_str))+str(my_str.count("e"))+"!" # 규칙3

    print(f"{url}의 비밀번호는 {password}입니다.")

url = "https://www.naver.com"
make_password(url)
url = "https://www.google.com"
make_password(url)
url = "https://www.daum.com"
make_password(url)
```

# 딕셔너리

```python
cabinet = {1:"a", 2:"b", 100:"c"}

print(cabinet[1]) # a
print(cabinet.get(1)) # a
print(cabinet.get(5)) # None
# print(cabinet[5]) # error
print(cabinet.get(5, "키 없음")) # 키가 없으면 뒤에 있는 문자열 실행

print(1 in cabinet) # True

cabinet[3] = "d"
print(cabinet)

cabinet[3] = "c"
print(cabinet)

del cabinet[1]
print(cabinet)

print(cabinet.keys())
print(cabinet.values())
print(cabinet.items())

cabinet.clear()
print(cabinet)
```

### 5-6. 퀴즈
댓글 이벤트를 할 예정이다. 1명은 치킨, 3명은 커피 쿠폰을 줄 예정인데, 추첨 프로그램을 작성해보자.

- 댓글은 20명이 작성했고 아이디는 1~20으로 가정하자
- 무작위로 추첨하되 중복 불가이다
- random 모듈의 shuffle, sample을 활용하자

```python
from random import *
comment_list = [1,2,3,4,5]
print(comment_list)
shuffle(comment_list)
print(comment_list)
print(sample(comment_list, 1))
```

```python
from random import *

comment_list = []
for i in range(1,21):
    comment_list.append(i)

print(comment_list)
shuffle(comment_list)
print(comment_list)

winner = sample(comment_list, 4)
print("치킨 당첨자는 {}입니다.".format(winner[:1]))
print("커피 당첨자는 {}입니다.".format(winner[1:4]))
```

```python
from random import *

users = range(1,21)
print(users, type(users))
users = list(users)
print(users)

shuffle(users)
print(users)

winners = sample(users, 4)
print("치킨 당첨자는 {0}입니다.".format(winners[0]))
print("커피 당첨자는 {0}입니다.".format(winners[1:4]))
```
