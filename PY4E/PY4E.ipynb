# C03_조건부 실행

```python
# 40시간이 넘어가면 초과근무 급여 적용하기
hours = float(input("몇 시간 일했어요? "))
rates = float(input("시급은 얼마에요? "))
if hours > 40:
    overtime = 40-hours
    total = (hours*rates)+(overtime*(0.5*rates)) # 40시간 오버타임이면 1.5배 추가 급여
else:
    total = hours * rates

print(f"임금은 총 {total}입니다.")
```

# C04_function

```python
def computepay(hours, rates):
    if hours > 40:
        overtime = 40-hours
        total = (hours*rates)+(overtime*(0.5*rates)) # 40시간 오버타임이면 1.5배 추가 급여
    else:
        total = hours * rates
    return total
    

hours = float(input("몇 시간 일했어요? "))
rates = float(input("시급은 얼마에요? "))

total = computepay(hours, rates)
print(f"임금은 총 {total}입니다.")
```

# C05_loop

```python
# input으로 들어오는 숫자를 합계, 개수, 평균 값 구하는 코드 만들기
num = 0
total = 0
while True:
    input_something = input('숫자 혹은 done을 입력하세요: ')
    if input_something == 'done':
        break
    try:
        input_num = float(input_something)
    except:
        print('잘못된 접근입니다. 숫자 혹은 done을 입력하세요.')
        continue
    
    num += 1
    total += input_num

print(f"합계는 {total}, 개수는 {num}, 평균은 {total/num}입니다.")
```
