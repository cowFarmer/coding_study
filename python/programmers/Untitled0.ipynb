```python
def solution(id_list, report, k):
    answer = [0]*len(id_list)
    report = list(set(report))
    report_count = []

    for i in range(len(report)):
        report_count.append(report[i].split()[1])
    
    # print(report_count)

    for i in range(len(id_list)):
        if report_count.count(id_list[i]) >= k:
            # report에서 value 값 갖고 있는 key += 1
            # print(id_list[i])
            for j in range(len(report)):
                if id_list[i] == report[j].split()[1]:
                    answer[id_list.index(report[j].split()[0])] += 1

    return answer

a = ["muzi", "frodo", "apeach", "neo"]
b = ["muzi frodo", "apeach frodo", "frodo neo", "muzi neo", "apeach muzi"]
c = 2
solution(a,b,c)
```
