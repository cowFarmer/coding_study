https://www.acmicpc.net/problem/10162

전자레인지

```python
time = int(input())
a = b = c = 0

if time%10 != 0:
    print(-1)
else:
    a = time//300
    b = time%300//60
    c = time%300%60//10
    print(a, b, c)
```

https://www.acmicpc.net/problem/2720

세탁소 사장 동혁

```python
n = int(input())
for i in range(n):
    cent = int(input())
    a = cent//25
    b = cent%25//10
    c = cent%25%10//5
    d = cent%25%10%5//1
    print(a, b, c, d)
```

https://www.acmicpc.net/problem/11034

캥거루 세마리2

```python
while True:
    try:
        a, b, c = map(int, input().split())
        print(max(b-a, c-b)-1)
    except:
        break
```
