https://codeup.kr/problem.php?id=6097

```python
array = []
w, h = map(int, input().split())
for width in range(w):
    height = [0]*h
    array.append(height)

n = int(input())
for i in range(n):
    l, d, x, y = map(int, input().split())
    x, y = x-1, y-1
    # 길이, 방향(0은 가로, 1은 세로), 좌표
    if d == 0:
        for j in range(l):
            array[x][y+j] = 1
    else:
        for j in range(l):
            array[x+j][y] = 1

for width in range(w):
    for height in range(h):
        print(array[width][height], end=' ')
    print()
```
