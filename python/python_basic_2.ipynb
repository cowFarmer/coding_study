# 6-6. 퀴즈
택시 기사이다.   
10명의 승객과 매칭 기회가 있을 때, 총 탑승 승객 수를 구해보자.

조건
- 승객별 운행 소요 시간은 5분~50분 사이의 수로 정해진다
- 5~15분 사이의 승객만 매칭해야 한다.

```python
from random import *

total = 0 # 총 탑승 가능한 승객

for i in range(1, 11):
    ride_time = randint(5,50)
    if ride_time >= 5 and ride_time <= 15:
        total += 1
        print(f"[O] {i}번째 손님(소요시간 : {ride_time}분)")
    else:
        print(f"[] {i}번째 손님(소요시간 : {ride_time}분)")
    

print(f"총 탑승 가능한 승객: {total}명")
```
