# C03_조건부 실행

````python
# 40시간이 넘어가면 초과근무 급여 적용하기
hours = float(input("몇 시간 일했어요? "))
rates = float(input("시급은 얼마에요? "))
if hours > 40:
    overtime = 40-hours
    total = (hours*rates)+(overtime*(0.5*rates)) # 40시간 오버타임이면 1.5배 추가 급여
else:
    total = hours * rates

print(f"임금은 총 {total}입니다.")
````

# C04_function

````python
def computepay(hours, rates):
    if hours > 40:
        overtime = 40-hours
        total = (hours*rates)+(overtime*(0.5*rates)) # 40시간 오버타임이면 1.5배 추가 급여
    else:
        total = hours * rates
    return total
    

hours = float(input("몇 시간 일했어요? "))
rates = float(input("시급은 얼마에요? "))

total = computepay(hours, rates)
print(f"임금은 총 {total}입니다.")
````

# C05_loop

````python
# input으로 들어오는 숫자를 합계, 개수, 평균 값 구하는 코드 만들기
num = 0
total = 0
while True:
    input_something = input('숫자 혹은 done을 입력하세요: ')
    if input_something == 'done':
        break
    try:
        input_num = float(input_something)
    except:
        print('잘못된 접근입니다. 숫자 혹은 done을 입력하세요.')
        continue
    
    num += 1
    total += input_num

print(f"합계는 {total}, 개수는 {num}, 평균은 {total/num}입니다.")
````

# C09_dict

딕셔너리를 이용해 이름 빈도수 세기
1. 이미 저장되어 있는 이름인지 확인한다.   
2-1. 만약 저장되어 있는 경우 1을 더한다.   
2-2. 저장되어 있지 않은 경우 이름을 저장하고 1을 배정한다.   
3. 최종 결과 중 가장 빈도가 많은 이름을 찾는다.

````python
counts = dict()
names = ['a', 'b', 'c', 'd', 'e', 'a', 'b', 'a', 'e', 'e', 'e', 'e']

# 1번
for name in names:
    if name in counts:
        counts[name] += 1
    else:
        counts[name] = 1
print(counts)

counts = dict()
# 2번 get() 사용하기
for name in names:
    counts[name] = counts.get(name, 0)+1
    # counts dictionary에 name이라는 key가 존재하는 경우 name의 value를 가져오고 아니면 키에 0을 추가하라는 의미이다.
print(counts)
````

````python
# name = input('Enter file:')
# handle = open(name)
handle = 'i have a lot of chicken and a this is test text and a\n this is test test test test'

counts = dict()
# 문자의 개수 새기
# for line in handle:
#     words = line.split()
#     for word in words:
#         counts[word] = counts.get(word,0) + 1

# 문자열 개수 새기
words = handle.split()
for word in words:
    counts[word] = counts.get(word,0)+1
print(words)
print(counts)

bigcount = None
bigword = None
for word,count in counts.items():
    if bigcount is None or count > bigcount:
        bigword = word
        bigcount = count

print(bigword, bigcount)
````

# C10_tuple

````python
# fname = input("Enter File: ")
# if len(fname) < 1 : fname = "clown.txt"
# hand = open(fname)
hand = 'i have a lot of chicken and a this is test text and a\n this is test test test test'

di = dict()
for lin in hand:
    lin = lin.rstrip()
    wds = lin.split()
    for w in wds:
        di[w] = di.get(w,0)+1

tmp = list()
for k,v in di.items():
    newt = (v,k)
    tmp.append(newt)

tmp = sorted(tmp, reverse = True)

for v,k in tmp[:5]:
    print(k,v)
````
