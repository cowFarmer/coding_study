```python
# arr_a = [1,2,3,3,3,3,4,4] # [4, 2]
arr_a = [3,2,4,4,2,5,2,5,5] # [3, 2, 3]
# arr_a = [3,5,7,9,1]       # [-1]


arr_a.sort()
temp_list = list(set(arr_a)) # 배열의 중복 제거
counter = 0
print_list = []

for i in range(len(temp_list)):
    if arr_a.count(temp_list[i]) != 1:
        counter += 1
        print_list.append(arr_a.count(temp_list[i]))
if counter == 0:
    print_list.append(-1)
print(print_list)
```
