# DFS

[백준 2606번 바이러스 문제풀기](https://www.acmicpc.net/problem/2606)

```python
count_com = int(input())
count_edge = int(input())

stack = []


for i in range(count_edge):
    stack.append(list(input().split()))

print(stack) # [['1', '2'], ['2', '3'], ['1', '5'], ['5', '2'], ['5', '6'], ['4', '7']]

print(stack[0])

def dfs(start):
    for i in range(start, count_com):
        
        count = 0
        print(count_com)
        for j in range(i, count_com):
            if stack[start][0] == stack[j][i]:
                count += 1
                dfs(stack[j])
    return count

dfs(0)
```
